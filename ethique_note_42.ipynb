## Column definitions for online_shoppers_intention dataset (detailed)

- **Administrative**: Number of pages visited by the user about account management.
- **Administrative_Duration**: Total time (in seconds) spent by the user on administrative pages.
- **Informational**: Number of pages visited by the user about information (e.g., site, product, or policy info).
- **Informational_Duration**: Total time (in seconds) spent by the user on informational pages.
- **ProductRelated**: Number of product-related pages visited by the user during the session.
- **ProductRelated_Duration**: Total time (in seconds) spent by the user on product-related pages.
- **BounceRates**: The percentage of visitors who enter the site and leave without continuing to other pages.
- **ExitRates**: The percentage of exits from the site that occurred from a given page.
- **PageValues**: Average value for a web page that a user visited before completing a transaction.
- **SpecialDay**: Indicates the closeness of the site visit to a special day (e.g., Valentine’s Day), normalized between 0 and 1.
- **Month**: Month of the year when the user visited the website (e.g., 'Feb', 'Mar', ...).
- **OperatingSystems**: Numeric code (1–8) denoting the operating system used by the visitor (as recorded by user-agent parsing).
- **Browser**: Numeric code (1–13) denoting the web browser used (e.g., Chrome, Firefox, Safari), as recorded by user-agent parsing.
- **Region**: Numeric code (1–9) denoting the geographic region of the visitor.
- **TrafficType**: Numeric code (1–20) denoting the type of traffic (e.g., direct, referral, paid, etc.).
- **VisitorType**: Type of visitor: 'New_Visitor', 'Returning_Visitor', or 'Other'.
- **Weekend**: Boolean indicating whether the visit was on a weekend (TRUE/FALSE).
- **Revenue**: Boolean indicating whether the user completed a purchase (TRUE/FALSE).

# Datasheet du jeu de données : online_shoppers_intention

**Description générale :**
Ce jeu de données contient des informations sur les comportements d'achat en ligne d'utilisateurs, incluant des variables numériques et catégorielles liées à la navigation, aux transactions, et à l'intention d'achat.

**Provenance :**
- From : (https://www.kaggle.com/datasets/adilshamim8/online?resource=download)
- Source : [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Online+Shoppers+Purchasing+Intention+Dataset)
- Collecte : Données collectées lors de sessions de navigation sur un site e-commerce.
- Période : 2018

**Objectif :**
- Prédire si un utilisateur finalisera un achat lors de sa session.
- Analyser les facteurs influençant l'intention d'achat en ligne.

**Colonnes principales :**
- Variables numériques : nombre de pages visitées, durée de session, montant estimé, etc.
- Variables catégorielles : type de visiteur, mois, région, etc.
- Variable cible : Revenue (achat réalisé ou non lors de la session).

**Taille :**
- Nombre d'observations : 12 330
- Nombre de variables : 18

**Utilisations typiques :**
- Analyse comportementale
- Modélisation prédictive (classification)
- Visualisation de tendances d'achat

**Limitations et biais potentiels :**
- Données issues d'un seul site e-commerce
- Possible déséquilibre de classes (plus de non-achats que d'achats)
- Données collectées sur une période limitée

**Éthique et confidentialité :**
- Données anonymisées, pas d'informations personnelles identifiables
- Respect du RGPD

**Références :**
- https://archive.ics.uci.edu/ml/datasets/Online+Shoppers+Purchasing+Intention+Dataset
- https://www.kaggle.com/code/devraai/online-shoppers-intent-analysis

# Action sur le dataset : 


```python
# Plot histograms for some continuous variables
continuous_features = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']

plt.figure(figsize=(15, 10))
for i, col in enumerate(continuous_features):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()

# Plot count plot for categorical variable 'Revenue'
plt.figure(figsize=(6, 4))
sns.countplot(x='Revenue', data=df, palette='Set2')
plt.title('Revenue Distribution')
plt.show()
```
